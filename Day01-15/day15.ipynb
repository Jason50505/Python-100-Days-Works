```python
from PIL import Image, ImageFilter
from matplotlib.pyplot import imshow
%matplotlib inline
```

```python
# open and show the image
img = Image.open('test.PNG')

# img.show()
imshow(img)
```

```python
# Information about the object
img.format, img.size, img.mode, img.filename, img.format_description, img.info
```

```python
box = (80, 260, 870, 560)
img2 = img.crop(box=box)
imshow(img2)
```

```python
size = 300, 400
img.thumbnail(size=size)
imshow(img)
img.size
```

```python
# Resize
w, h = img2.size
img3 = img2.resize(size=(int(w / 1.5), int(h / 3)))
imshow(img3)
```

```python
# Paste
img = Image.open('test.PNG')
img.paste(img3, box=(0, 90))
imshow(img)
```

```python
# Rotate
img4 = img.rotate(-70)
imshow(img4)
```

```python
# Transpose
img5 = img.transpose(Image.ROTATE_180)
imshow(img5)
```

```python
# Modify the pixels
img = Image.open('test.PNG')

for x in range(80, 870):
    for y in range(260, 560):
        img.putpixel((x, y), (0, 0, 0))

imshow(img)
```

```python
# Filter
img = Image.open('test.PNG')
img6 = img.filter(ImageFilter.BLUR)
imshow(img6)
```
